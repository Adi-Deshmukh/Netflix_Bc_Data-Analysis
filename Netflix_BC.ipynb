# Data Cleaning

```python
# first split the data from string to list then use explode

df
```

```python

# Custom function to split and clean cast values
def split_values(x):
    if isinstance(x, str) and x.strip() != '':
        parts = x.split(', ')
        # Remove the actual NaN or empty parts from within the list
        parts = [p for p in parts if pd.notna(p) and p.strip() != '']
        return parts
    else:
        return []  # Return empty list if x is not a valid non-empty string
```

```python
df['cast'] = df['cast'].fillna('') # fills the NaN with''(empty sapce so that the above func runs soomthly as we are checking with '' in the func)
```

```python
df["cast"] = df["cast"].apply(split_values)
# we will change [] values to NaN before exploding

df["cast"] = df["cast"].apply(lambda x: np.nan if  isinstance(x,list) and len(x)==0 else x)
df["cast"]
```

```python
df["cast"].isna().sum()
```

```python
# now we explode cast column
df_c_exp = df.explode("cast").reset_index(drop=True) 
```

```python
df_c_exp["director"] = df_c_exp["director"].apply(split_values)
df_cd_exp = df_c_exp.explode("director").reset_index(drop=True)
```

```python
df_cd_exp.drop_duplicates(keep="first", inplace=True)
```

```python
df_cd_exp["country"]= df_cd_exp["country"].apply(split_values)
df_cdC_exp= df_cd_exp.explode("country").reset_index(drop=True) #--> make explode column as a custom function
```

```python
df_cdC_exp["listed_in"] = df_cdC_exp["listed_in"].apply(split_values)
df_cdCl_exp = df_cdC_exp.explode("listed_in").reset_index(drop=True)
df_cdCl_exp
```

```python
# Lets first impute the missing values from movies

# change this into a custom function later



def get_groupwise_mode(df, group_cols, target_col, mode_col_name = "mode_value" ):   # converts the list into string
         
    df_mode= (
        #first drop the Nan values to find the correct mode to impute
        df_movies.dropna(subset= target_col)
        .groupby(group_cols)[target_col]
        
         #this if director mode is empty oeds not check for iloc[0] (iloc[0] for not multiple nide)
        .agg(lambda x:x.mode().iloc[0] if not x.mode().empty else np.nan) 
        .reset_index()
        .rename(columns={target_col:mode_col_name})
    )
    return df_mode



def multilevel_impute(df, target_col, group_levels):
    df_copy = df.copy()
    for i, level in enumerate(group_levels):
        print(f"Level {i+1}: Imputing '{target_col}' using group {level or 'Unknown'}")
        mode_col = f"{target_col}_mode_lvl{i+1}"

        # Compute mode values for the grouping level
       
        if level:
            mode_df = get_groupwise_mode(df_copy, level, target_col, mode_col_name=mode_col)
            df_copy = df_copy.merge(mode_df, on=level, how="left")

            # Fill missing values using mode
            df_copy[target_col] = df_copy[target_col].fillna(df_copy[mode_col])

            # Drop the temporary mode column
            df_copy.drop(columns=[mode_col], inplace=True)

        else:
            # Fill any remaining nulls with 'Unknown' placeholder
            df_copy[target_col] = df_copy[target_col].fillna(f"Unknown {target_col.title()}")

        # Stop if all NaNs are filled
        if df_copy[target_col].isna().sum() == 0:
            print(f" All missing values in '{target_col}' imputed.\n")
            break
    return df_copy



# only difference from the multiple_impute was the if statement and the loop

def multiple_impute(df, impute_plan):
    df_copy = df.copy()

    for step in impute_plan:
        target_col = step['target']
        group_levels = step['group_levels']
        df_copy = multilevel_impute(df_copy, target_col, group_levels)


    return df_copy
    

```

```python
impute_plan = [
    {
        "target": "director",
        "group_levels": [
            ["country", "cast", "listed_in"],
            ["country", "cast"],
            []  # global mode
        ]
    },
    {
        "target": "cast",
        "group_levels": [
            ["country", "director", "listed_in"],
            ["country", "director"],
            []
        ]
    },
    {
        "target": "country",
        "group_levels": [
            ["director", "cast"],
            []
        ]
    },
     {
        "target": "rating",
        "group_levels": [
            ["cast", "director", "listed_in"],
            ["listed_in", "director"],
            []
        ]
    }
]

df_movies_cleaned = multiple_impute(df_movies, impute_plan)
df_movies_cleaned.isna().sum()
```

```python
# to clean duration column

median = df_movies_cleaned['duration'].str.extract('(\d+)').astype(float).median().iloc[0]
df_movies_cleaned['duration'] =  df_movies_cleaned['duration'].fillna(f"{int(median)} min")
```
